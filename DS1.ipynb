```python
import numpy as np
import pandas as pd
np.random.seed(52)

data={
    "x" : np.arange(1,8),
    "y" : np.random.randint(20,50,size=7)
}
df=pd.DataFrame(data)
print(df)
```

```python

import matplotlib.pyplot as plt
plt.figure(figsize = (3,3))
x = df["x"]
y = df["y"]
plt.plot(x,y,marker="*",color='violet')
plt.title("AMPLIFIER")
plt.xlabel("amplitude")
plt.ylabel("frequency")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(1, 6)

plt.figure(figsize=(5,5))

plt.subplot(2, 2, 1)
plt.plot(x, x)
plt.title("Line Plot")

plt.subplot(2, 2, 2)
plt.bar(x, x*2)
plt.title("Bar Chart")

plt.subplot(2, 2, 3)
plt.scatter(x, x*3)
plt.title("Scatter Plot")

plt.subplot(2, 2, 4)
plt.pie(x, labels=x)
plt.title("Pie Chart")

plt.tight_layout()
plt.show()
```
